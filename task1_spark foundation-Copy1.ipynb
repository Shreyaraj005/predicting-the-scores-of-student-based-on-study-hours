```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv("http://bit.ly/w-data")
df.head(8)
```

```python
#scatter plot
plt.scatter(x=df.Hours,y=df.Scores)
plt.title('scatter plot'),plt.xlabel('hours'),plt.xlabel('hours')
plt.ylabel('score')
```

```python
#correlation
df.corr()
```

```python
#visualisation with seaborn
import seaborn as sns
sns.pairplot(df)
```

```python
#independent and dependent feature
x=df[['Hours']]
x
```

```python
y=df['Scores']
np.array(x).shape
```

```python
#train test split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=0)
```

```python
#apply simple linear regression
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
print('coffecient or slop:',regression.coef_)
print('intercept:',regression.intercept_)
```

```python
#plot raining data plot best fit line
plt.scatter(x_train,y_train)
plt.plot(x_train,regression.predict(x_train))
plt.title('best fit line')
plt.xlabel('hours')
plt.ylabel('scores')
```

here the graph shows the best fit line for the given data



```python
#ols prediction
import statsmodels.api as sm
```

```python
model=sm.OLS(x_train,y_train).fit()
```

```python
prediction=model.predict(x_test)
print (prediction)
```

```python
print(model.summary())
```

```python
#prediction for new data
print('predicted score:',regression.predict([[9.25]]))
```
